```python
# test for a single dataset
dataset_name = 'adult'
clean_df = clean_df[clean_df['dataset_name'] == dataset_name]
model_df = clean_df.drop(columns='dataset_name')
```

```python
target = 'final_val_balanced_accuracy'
X = model_df.drop(target, axis=1)
y = model_df[target]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
import naiveautoml

naml = naiveautoml.NaiveAutoML(timeout_overall=60, show_progress=True)
naml.fit(X_train, y_train)
```
